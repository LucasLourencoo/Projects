```python
class selec_cor:
    OK = '\033[92m' #VERDE
    WARNING = '\033[93m' #AMARELO
    FAIL = '\033[91m' #VERMELHO
    RESET = '\033[0m' #RESETAR CORES
    REVERSE = "\033[;7m" #COR REVERSA
    DETAIL = "\033[1;34m" #AZUL
    BOLD = "\033[;1m" #NEGRITO


meta_media = int(70)
nmr_sec = str('(21) 99999-9999') #número da secretaria
reqde_certify = 95 #requisito de certificado de reconhecimento de melhor aluno


nome_procurador = input('Favor, inserir o seu nome.').strip().title()

if 'Lucas' or 'Matheus' in nome_procurador:
    print('Nome >{}< validado.\nProsseguindo...'.format(nome_procurador))
else:
    print( selec_cor.FAIL + 'Error N!' + selec_cor.RESET + '\nO dado inserido não é válido. Favor, preencher corretamente.')


#cpf do procurador
cpfdo_procurador = input('Insira seu CPF: ').strip()

ponto = '.'
traco = '-'
contador = 10
posicao = 0
acumulo = 0
resto = 0
cont = 11
pos = 0
res = 0
acu = 0


#operação para validar o CPF
if len(cpfdo_procurador) == 11 or cpfdo_procurador.isnumeric():
    cpfdo_procurador_formatado = cpfdo_procurador[0:3] + ponto + cpfdo_procurador[3:6] + ponto + cpfdo_procurador[6:9] + traco + cpfdo_procurador[9:]
    print(f'O CPF informado foi: {cpfdo_procurador_formatado}')

    for i in range(len(cpfdo_procurador[0:9])):
        primeiro = int(cpfdo_procurador[posicao]) * contador
        posicao += 1
        contador -= 1  
        acumulo = acumulo + primeiro

        if contador < 2 and posicao > 8:
            resto = acumulo % 11
            primeiro_verificador = 11 - resto
            if primeiro_verificador >= 10:
                primeiro_verificador = 0
            print(f'Este é o primeiro dígito verificador de seu CPF: {primeiro_verificador}')

    
    for x in range(len(cpfdo_procurador[0:10])):
        segundo = int(cpfdo_procurador[pos]) * cont
        pos += 1
        cont -= 1  
        acu = acu + segundo

        if cont < 2 and pos > 9:
            res = acu % 11
            segundo_verificador = 11 - res
            if segundo_verificador >= 10:
                segundo_verificador = 0
            print(f'Este é o segundo dígito verificador de seu CPF: {segundo_verificador}')

    primeiro_verificador = str(primeiro_verificador)
    segundo_verificador = str(segundo_verificador)

    if primeiro_verificador == cpfdo_procurador[9] and segundo_verificador == cpfdo_procurador[10]:
        print('O CPF foi analisado e aprovado. CPF válido!')

    else:
        print('CPF inválido!')


elif len(cpfdo_procurador) == 14 and '.' == cpfdo_procurador[3] and '.' == cpfdo_procurador[7] and '-' == cpfdo_procurador[11]:
    print(f'O CPF informado foi: {cpfdo_procurador}')
    cpfdo_procurador = cpfdo_procurador.replace('.', '')
    cpfdo_procurador = cpfdo_procurador.replace('-', '')

    for i in range(len(cpfdo_procurador[0:9])):
        primeiro = int(cpfdo_procurador[posicao]) * contador
        posicao += 1
        contador -= 1  
        acumulo = acumulo + primeiro

        if contador < 2 and posicao > 8:
            resto = acumulo % 11
            primeiro_verificador = 11 - resto
            if primeiro_verificador >= 10:
                primeiro_verificador = 0
            print(f'Este é o primeiro dígito verificador de seu CPF: {primeiro_verificador}')

    
    for x in range(len(cpfdo_procurador[0:10])):
        segundo = int(cpfdo_procurador[pos]) * cont
        pos += 1
        cont -= 1  
        acu = acu + segundo

        if cont < 2 and pos > 9:
            res = acu % 11
            segundo_verificador = 11 - res
            if segundo_verificador >= 10:
                segundo_verificador = 0
            print(f'Este é o segundo dígito verificador de seu CPF: {segundo_verificador}')

    primeiro_verificador = str(primeiro_verificador)
    segundo_verificador = str(segundo_verificador)

    if primeiro_verificador == cpfdo_procurador[9] and segundo_verificador == cpfdo_procurador[10]:
        print('O CPF foi analisado e aprovado. CPF válido!')
    else:
        print('CPF inválido!')
else:
    print('CPF inválido!')



#nome do aluno
nome_aluno = str(input('Informe o nome do aluno: \n')).strip().title()

if nome_aluno.isalpha:
    print('Insira os dados do aluno(a) {} a seguir:'.format(nome_aluno))
else:
    print( selec_cor.FAIL + 'Error N!' + selec_cor.RESET + '\nO dado inserido não é válido. Favor, preencher corretamente.')



#curso que o aluno está fazendo
curso = str(input('Qual o curso do aluno {} está querendo saber a média? \n'
.format(nome_aluno))).strip().title()

if len(curso) <= 3:
    curso = curso.upper()
    print('O curso do aluno(a) {} é'.format(nome_aluno) + selec_cor.WARNING + ' {}'.format(curso) + selec_cor.RESET)

elif curso.isalpha:
    print('O curso do aluno(a) {} é'.format(nome_aluno) + selec_cor.WARNING + ' {}'.format(curso) + selec_cor.RESET)

else:
    print( selec_cor.FAIL + 'Error C!' + selec_cor.RESET + '\nO dado inserido não é válido. Favor, preencher corretamente.')



#quantidade de aulas do curso inserido anteriormente 
qnt_aula = int(input('Insira a quantidade de aulas que o curso de {} tem: \n'.format(curso)).strip())
if qnt_aula <= 36:
    print('A quantidade de aulas do curso de {} é de {} aulas'.format(curso, qnt_aula))
else:
    print( selec_cor.FAIL + 'Error Q!' + selec_cor.RESET + '\nO dado inserido não é válido. Favor, preencher corretamente.')
    


if nome_aluno and curso and qnt_aula:

    #inserir notas
    contadorwhile = 0
    soma = 0
    while contadorwhile < int(qnt_aula):
        qnt_aula2 = int(qnt_aula) - contadorwhile
        soma += float(input('Insira a nota que o aluno(a) {a} tirou correspondente à AULA {c}.\n(Total de aulas restantes a serem preenchidas: {q})\n'
        .format(q = qnt_aula2, c = contadorwhile + 1, a = nome_aluno)))
        contadorwhile += 1

    media_final = soma / int(qnt_aula)


    #Media maior que meta
    if  meta_media < media_final >= meta_media:
        print( selec_cor.OK + 'O aluno(a) {} passou no curso de {}!\nSua média foi de {}'
        .format(nome_aluno, curso, media_final) + selec_cor.RESET) 
        
        #media maior do que requisito pra certificado 
        if media_final > reqde_certify:
            print( selec_cor.OK + 'O aluno {} tirou uma nota maior do que {}.\nPor este motivo, o mesmo estará recebendo um certificado de reconhecimento.\nFavor, entrar em contato com a secretaria através do número {}.'
            .format(nome_aluno, reqde_certify, nmr_sec) + selec_cor.RESET)
        else:
            pass


    #media igual à meta
    elif media_final == meta_media:
        print( selec_cor.WARNING + 'O aluno {} passou exatamente com a nota mínima.'.format(nome_aluno) + selec_cor.RESET)


    #media abaixo da meta
    else:
        print( selec_cor.FAIL +  'O aluno(a) {} não passou no curso de {}. Sua média foi {}'
        .format(nome_aluno, curso, media_final) + selec_cor.RESET)
        
        #muito próximo à aprovação
        if 69.99 > media_final >= 65:
            print( 'O aluno {} tirou uma nota muito próxima da aprovação\nFavor, entrar em contato com a secretaria através do número {}.'
            .format(nome_aluno, nmr_sec))
        else: 
            pass

else:
    print('Não foi possível concluir a operação.\nFavor, preencher todos os dados corretamente.')
```
